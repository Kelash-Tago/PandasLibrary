```python
#clean data 

with open("data.json","r") as file:
    data=json.load(file)
clean_users=[] 
seen_users=set()
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    
    for user in data:
        raw_rating=user.get("rating").strip().lower()
        if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
        user["rating"]=raw_rating
        

        #for age some user have not age key assigning None to them
        raw_age=user.get("age")
        if(raw_age==None):
            user["age"]=None

        #for unique users
        
        if user["name"].strip() in seen_users:
            continue
        clean_users.append(user)
        seen_users.add(user["name"].strip())
            
    return clean_users       

#callig clean data function
clean_data(data)  
```
