```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Student_ID": [1, 2, 3],
    "Name": ["Ali", "Sara", "Ahmed"],
    "Math": [85, 90, 78],
    "Science": [88, 92, 80],
    "English": [75, 85, 82]
})
```

```python
df
```

```python
melted_data=df.melt(
    id_vars=["Student_ID","Name"],
    var_name="Subjects",
    value_name="Marks"
)
```

```python
melted_data
```

```python
melted_data.pivot(
    index=["Student_ID","Name"],
    columns="Subjects",
    values="Marks"
)
```
