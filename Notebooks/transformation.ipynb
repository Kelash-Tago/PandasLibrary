```python
import pandas as pd
```

```python
df=pd.read_csv("employee_data.csv")
```

```python
df["tax"]=df["Salary"].apply(lambda x: "25%" if x>50000 else "15%")
```

```python
# map function
map_value={"IT":"Information Technology","HR":"Human Resources","Finance":"Finance","Marketing":"Marketing"}
df["Department"]=df["Department"].map(map_value)
```

```python
# assign function

df.assign(new_salary=df["Salary"]*1.1)
```

```python
# replace function
df["Department"].replace("Information Technology","IT")
```
